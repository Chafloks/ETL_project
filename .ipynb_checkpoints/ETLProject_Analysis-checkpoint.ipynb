```python
# Dependencies and setup
import pandas as pd
import csv
```

```python
# Data files
region_data = "Resources/All_homes.csv"
score_data = "Resources/data_ACT-Data-by-District-2013-14.csv"
```

```python
# Convert to region_data to df
region_df = pd.read_csv(region_data, index_col = 0)
```

```python
# Filter region_df by Tennessee towns only
TN_towns = region_df.loc[region_df["State"] == "TN"].reset_index()
TN_towns.head()
```

```python
# Convert to score_data to df & format
score_df = pd.read_csv(score_data, header = [2,3])
score_df.columns = score_df.columns.map('_'.join)
score_df.rename(columns={'Unnamed: 0_level_0_System':'Region'}, inplace=True )
score_df["Region"] = score_df["Region"].str.title()
score_df.head()
```
